```python
playstore.drop(playstore.iloc[:, 9:], inplace=True, axis=1)
playstore.head(10)  # dropping columns that are unnecessary for this analysis
```

```python
playstore.drop_duplicates(inplace=True) # removing entries that are completely same for each column, if any value in any column is different then it won't be removed
```

```python
playstore.info()  # here we can see that 10841 - 10356 = 485 entries were removed.
                    # also important to note that "rating" column has the most missing values and "type" and "content rating" columns have only 1 missing value
```

```python
playstore.describe() # this shows only numerical columns and some statistics of it. here we can see that there are values over 5.0 which is not good.
```

```python
playstore.loc[playstore['Rating'] == 19] # here is obvious that this row has shifted values because there is missing value for Category. 
                                         # Since this is the only entry of this kind, it will be dropped from dataset
```

```python
playstore.drop(playstore.index[(playstore['Rating'] == 19)],axis=0, inplace=True) #removing row with shifted values
playstore.describe()
```

```python
playstore.groupby('Category').mean() # this code is to check how to fill missing values in dataset for Rating column. based on this it will be good to 
                                     # fill the missing data with average ratings for each category separately
```

```python
playstore.isna().sum() # number of missing values before filling
```

```python
playstore['Rating'] = playstore.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.mean()))
playstore['Rating'] = playstore['Rating'].round(1)   # filling missing values and rounding rating to 1 decimal places so it matches original data
```

```python
playstore.isna().sum() # here we can see there is no more missing values in Rating column.
```

```python
categorical = playstore.dtypes[playstore.dtypes == 'object'].index
playstore[categorical].describe()
```

```python
playstore.head()
```

```python
playstore['Reviews'] = playstore['Reviews'].astype(float) # changing dtype from object to float because Reviews column represents number of reviews
```

```python
playstore['Size'].replace(to_replace='Varies with device', method='ffill', inplace=True) # replacing "Varies with device" by any previous value in Size column
```

```python
playstore['Numerical_part'] = playstore.Size.str[:-1] # making temporary columns to help with data manipulation
playstore['String_value'] = playstore.Size.str[-1]
```

```python
playstore.head()
```

```python
categorical = playstore.dtypes[playstore.dtypes == 'object'].index # seeing there are only 2 unique values for new column
playstore[categorical].describe()
```

```python
playstore.loc[playstore['String_value'] != 'M'] # checking to see which is other value for new column
```

```python
playstore['Numerical_part'] = playstore['Numerical_part'].astype(float) # making new column float type
```

```python
playstore.loc[playstore['String_value'] == 'k', 'Numerical_part'] = playstore['Numerical_part'] * 0.001 # transforming K into M and leaving M as main unit
```

```python
playstore['Numerical_part'] = playstore['Numerical_part'].round(1) # rounding size to 1 decimal places since original data is only 1 decimal place
```

```python
playstore['Size'] = playstore['Numerical_part'].astype(str) + 'M' # making Size column consistent with units
```

```python
playstore['Size[M]'] = playstore['Numerical_part'] # making new column that enables numerical analisys on Size of apps
```

```python
playstore.drop(['Numerical_part', 'String_value'], axis=1, inplace=True) # deleting temporary columns
```

```python
playstore = playstore[['App', 'Category', 'Rating', 'Reviews', 'Size', 'Size[M]', 'Installs', 'Type', 'Price', 'Content Rating']] # rearranging dataset
```

```python
# playstore['Installs'] = playstore.Installs.str[:-1]
```

```python
playstore['Installs'] = playstore['Installs'].str.rstrip('+') # deleting "+" sign
```

```python
playstore['Installs'] = playstore.Installs.str.replace(',', '').astype(float) # making Installs column float type and with this finishing with Installs column
```

```python
playstore.isna().sum() # looking for NAN values and there is 1 for Type column
```

```python
playstore.loc[playstore['Price'] == '0', 'Type'] = 'Free' # since Type column value: "Free" was the most common value in dataset, this was simple try to see if the missing value has price = 0
                                                          # in Price column, and it worked which can be seen in next step
```

```python
playstore.isna().sum() # seeing that there are no more NAN values. Type column is finished since there is only 2 unique values in this column (free and paid) and no NAN values.
```

```python
playstore['Price[$]'] = playstore['Price'].str.lstrip('$').astype(float).round(2) # making price duplicate column which will be float type so numerical analisys can be done on Price column
```

```python
playstore['Price'] = '$' + playstore['Price[$]'].astype(str) # fixing original Price column so its consistent with its values
```

```python
playstore = playstore[['App', 'Category', 'Rating', 'Reviews', 'Size', 'Size[M]', 'Installs', 'Type', 'Price', 'Price[$]', 'Content Rating']]
```

```python
playstore.loc[(playstore['Content Rating'] != 'Everyone') & (playstore['Content Rating'] != 'Teen') & (playstore['Content Rating'] != 'Everyone 10+') & (playstore['Content Rating'] != 'Mature 17+')]
# doing this process of elimination I found there are 2 rows with Content Rating = Unrated
```

```python
playstore.loc[playstore['Category'] == 'FAMILY', 'Content Rating'].describe()
```

```python
playstore.loc[playstore['Category'] == 'TOOLS', 'Content Rating'].describe() # by seeing this and previous step, conclusion is that the majority of Content Rating for specific Category is "Everyone"
```

```python
playstore['Content Rating'].replace(to_replace='Unrated', value='Everyone', inplace=True) # replacing "Unrated" value with "Everyone" and with this column Content rating is also clean.
```
