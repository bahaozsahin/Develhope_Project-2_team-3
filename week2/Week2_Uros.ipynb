```python
import pandas as pd
week2 = pd.read_csv(r'cleaned_playstore.csv')
```

Task 1: print (table) the unique names of all categories

```python
table = pd.DataFrame(week2['Category'].unique(), columns=['Unique Categories'])
table.head(50)
```

Task 2: plot a bar chart for categories with the total number of installing numbers in each category 

```python
import matplotlib.pyplot as plt
```

```python
task2 = week2.groupby(by='Category')['Installs'].sum()
```

```python
task2.plot.barh(xlabel = 'Number of Installs * 10^10', title = 'Number of App installs per category')
plt.show()
```

Task 3: plot a bar chart for the total prices of each paid app in each category ( the summation of all prices in the same category)

```python
x = week2.loc[week2['Type'] == 'Paid']
task3 = x.groupby(by='Category')['Price'].sum()
```

```python
task3.plot.barh(xlabel = 'total price [$]', title = 'Total app price for each category')
plt.show()
```

Task 4: plot a bar chart of the total profit of each category by multiplying the price by the number of installs

```python
x['Profit'] = x['Price'] * x['Installs']
```

```python
task4 = x.groupby(by='Category')['Profit'].sum()
```

```python
task4.plot.barh(xlabel = 'Apps profit * 10^8 [$]', title = 'Total profit of each category')
plt.show()
```
