```python
# tasks for this week :
# 1-    print (table) the unique names of all categories
# 2-    plot a bar chart for categories with the total number of installing numbers in each category 
# 3-    plot a bar chart for the total prices of each paid app in each category ( the summation of all prices in the same category)
# 4-    plot a bar chart of the total profit of each category by multiplying the price by the number of installs

import pandas as pd

cleaned_playstore = pd.read_csv('cleaned_playstore.csv', index_col=0)
cleaned_playstore_reviews = pd.read_csv('cleaned_playstore_reviews.csv', index_col=0)
```

```python
# 1

category_df = pd.DataFrame(cleaned_playstore.Category.unique(), columns=['Category'])
category_df
```

```python
# 2

import seaborn as sns
import matplotlib.pyplot as plt

category_installs_df = cleaned_playstore.groupby('Category').Installs.sum().reset_index()
sns.barplot(data = category_installs_df.sort_values(by = 'Installs', ascending = False), x='Category', y='Installs')

plt.xticks(rotation=90)
plt.title('Categories By Installs')
plt.show()
```

```python
# 3

category_pricesum_df = cleaned_playstore.groupby('Category').Price.sum().reset_index()
sns.barplot(data = category_pricesum_df.sort_values(by = 'Price', ascending = False), x='Category', y='Price')

plt.xticks(rotation=90)
plt.title('Categories By Sum of Paid App Prices')
plt.ylabel('Sum of Paid App Prices')
plt.show()
```

```python
# 4
cleaned_playstore['Profit'] = cleaned_playstore.Installs * cleaned_playstore.Price
category_profit_df = cleaned_playstore.groupby('Category').Profit.sum().reset_index()
sns.barplot(data = category_profit_df.sort_values(by = 'Profit', ascending = False), x='Category', y='Profit')

plt.xticks(rotation=90)
plt.title('Categories By Profit')
plt.show()
```
