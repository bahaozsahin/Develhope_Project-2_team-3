Find top 5 apps in each category by number of installs but separated by type of apps (Free and Paid)

```python
import pandas as pd
```

```python
data = pd.read_csv('../week2/cleaned_playstore.csv', index_col=0)
data.head()
```

```python
df = data.sort_values('Installs', ascending=False).groupby(['Category', 'Type']).head(5)
df
```

```python
pd.set_option('display.max_rows', 10000) # just to be able to see whole output (10000 is overkill but just to be sure none of the rows is missing)
```

```python
df.set_index(df['App']).groupby(['Category', 'Type'])['Installs'].nlargest(5).reset_index(2).head(10000)
```
